## Data analysis

### Subtask:
Analyze the data to gain deeper insights.


**Reasoning**:
Calculate the correlation matrix, group data by the target variable, and analyze numerical and categorical features as well as the 'PAY_' columns to gain deeper insights into the data.



```python
import pandas as pd

# Assuming you already have a DataFrame named df
# Example:
# df = pd.read_csv('your_file.csv')

# 1. Data Types
print("Data Types:\n", df.dtypes)

# 2. Missing Values
print("\nMissing Values:\n", df.isnull().sum())

# 3. Descriptive Statistics for Numerical Features
numerical_features = df.select_dtypes(include=['number'])
print("\nDescriptive Statistics for Numerical Features:\n", numerical_features.describe())

# 4. Value Frequencies for Categorical Features
categorical_features = df.select_dtypes(exclude=['number'])
for col in categorical_features.columns:
    print(f"\nValue Frequencies for {col}:\n{categorical_features[col].value_counts()}")

# 5. Shape of the Data
print(f"\nShape of the Data: {df.shape}")

# 6. Unique Values
for col in df.columns:
    print(f"\nUnique values for {col}:\n{df[col].unique()}")
```

```python
import pandas as pd
import numpy as np

# Calculate the correlation matrix
correlation_matrix = df.corr()
display(correlation_matrix)

# Identify highly correlated features
threshold = 0.7
highly_correlated_features = correlation_matrix[((correlation_matrix > threshold) | (correlation_matrix < -threshold)) & (correlation_matrix != 1.0)]
display(highly_correlated_features)

# Group data by 'default.payment.next.month' and calculate descriptive statistics
grouped_data = df.groupby('default.payment.next.month').agg(['mean', 'median', 'std'])
display(grouped_data)


# Analyze 'PAY_' columns
pay_cols = [col for col in df.columns if 'PAY_' in col]
for col in pay_cols:
    print(f"\nAnalysis for {col}:")
    default_rates = df.groupby(col)['default.payment.next.month'].mean()
    print(default_rates)

# Analyze categorical features (SEX, EDUCATION, MARRIAGE)
categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE']
for col in categorical_cols:
    print(f"\nAnalysis for {col}:")
    default_rates = df.groupby(col)['default.payment.next.month'].mean()
    print(default_rates)
```

## Data visualization

### Subtask:
Visualize key aspects of the data to gain further insights.


**Reasoning**:
Generate visualizations to understand the data better, as per the instructions. This includes histograms, box plots, scatter plots, a heatmap for the correlation matrix, and bar charts for categorical features.



```python
import os
import seaborn as sns
import matplotlib.pyplot as plt

# â†’ Define your features up top
features_to_plot = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1', 'BILL_AMT2', 'PAY_AMT2']
categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE']
os.makedirs('plots', exist_ok=True)

# Create 'plots' folder if it doesn't exist
os.makedirs('plots', exist_ok=True)

# Histograms
for i, col in enumerate(features_to_plot, 1):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    save_path = f"plots/hist_{col}.png"
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

# Box Plots
for col in features_to_plot:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    save_path = f"plots/box_{col}.png"
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

# Scatter Plots
scatter_specs = [
    ('LIMIT_BAL', 'BILL_AMT1'),
    ('LIMIT_BAL', 'PAY_AMT1'),
    ('BILL_AMT1', 'PAY_AMT1'),
    ('AGE', 'LIMIT_BAL')
]
for x, y in scatter_specs:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x, y=y, hue='default.payment.next.month', data=df)
    plt.title(f'{x} vs {y}')
    save_path = f"plots/scatter_{x}_vs_{y}.png"
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

# Heatmap of correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
save_path = "plots/heatmap_corr.png"
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.close()

# Bar Charts for Categorical Features
for col in categorical_cols:
    default_rates = df.groupby(col)['default.payment.next.month'].mean()
    plt.figure(figsize=(8, 6))
    sns.barplot(x=default_rates.index, y=default_rates.values)
    plt.title(f'Default Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('Default Rate')
    plt.xticks(rotation=45)
    save_path = f"plots/bar_{col}.png"
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

```

```python
from flask import Flask, render_template, send_file
from matplotlib.figure import Figure
import seaborn as sns
import pandas as pd
import os
from pyngrok import ngrok

# Load your dataset here (update the path if needed)
# Changed the filename from "your_data.csv" to "UCI_Credit_Card.csv"
df = pd.read_csv("UCI_Credit_Card.csv")  # Make sure this file exists in your Colab environment

app = Flask(__name__)
os.makedirs('static/plots', exist_ok=True)

features_to_plot = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1', 'BILL_AMT2', 'PAY_AMT2']
categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE']

def create_and_save_plot(plotfname, plot_func):
    fig = Figure(figsize=(8, 6), dpi=300)
    ax = fig.subplots()
    plot_func(ax)
    path = os.path.join('static/plots', plotfname)
    fig.savefig(path, bbox_inches='tight')
    return path

def generate_all_plots():
    plot_paths = []

    # Histograms
    for col in features_to_plot:
        def hist(ax, c=col):
            sns.histplot(df[c], kde=True, ax=ax)
            ax.set_title(f'Distribution of {c}')
        plot_paths.append(create_and_save_plot(f"hist_{col}.png", hist))

    # Box Plots
    for col in features_to_plot:
        def box(ax, c=col):
            sns.boxplot(x=df[c], ax=ax)
            ax.set_title(f'Box Plot of {c}')
        plot_paths.append(create_and_save_plot(f"box_{col}.png", box))

    # Scatter Plots
    scatter_pairs = [
        ('LIMIT_BAL', 'BILL_AMT1'),
        ('LIMIT_BAL', 'PAY_AMT1'),
        ('BILL_AMT1', 'PAY_AMT1'),
        ('AGE', 'LIMIT_BAL'),
    ]
    for x, y in scatter_pairs:
        def scat(ax, x=x, y=y):
            if 'default.payment.next.month' in df.columns:
                sns.scatterplot(x=x, y=y, hue='default.payment.next.month', data=df, ax=ax)
            else:
                sns.scatterplot(x=x, y=y, data=df, ax=ax)
            ax.set_title(f'{x} vs {y}')
        plot_paths.append(create_and_save_plot(f"scatter_{x}_vs_{y}.png", scat))

    # Heatmap
    def heat(ax):
        if 'default.payment.next.month' in df.columns:
            sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', ax=ax)
            ax.set_title('Correlation Matrix')
        else:
            ax.text(0.5, 0.5, "Correlation matrix not available due to missing 'default.payment.next.month'",
                    horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
            ax.set_title('Correlation Matrix')
    plot_paths.append(create_and_save_plot("heatmap_corr.png", heat))

    # Bar Charts
    for col in categorical_cols:
        if col in df.columns and 'default.payment.next.month' in df.columns:
            def bar(ax, c=col):
                rates = df.groupby(c)['default.payment.next.month'].mean()
                sns.barplot(x=rates.index, y=rates.values, ax=ax)
                ax.set_title(f'Default Rate by {c}')
                ax.set_xlabel(c)
                ax.set_ylabel('Default Rate')
                ax.tick_params(axis='x', rotation=45)
            plot_paths.append(create_and_save_plot(f"bar_{col}.png", bar))
        else:
            print(f"Skipping bar chart for {col}: Column not found or 'default.payment.next.month' missing.")


    return [f"/plots/{os.path.basename(p)}" for p in plot_paths]

@app.route('/')
def index():
    plot_urls = generate_all_plots()
    return render_template('index.html', plot_urls=plot_urls)

@app.route('/plots/<plotname>')
def serve_plot(plotname):
    return send_file(os.path.join('static/plots', plotname), mimetype='image/png')

# Start Flask with ngrok
if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print("ðŸŒ Public URL:", public_url)
    app.run(port=5000)
```
